## N Caminantes Aleatorios y guardan T pasos

Queremos un arreglo de NxT, de tal manera que podramos analizar sus datos.

```julia
function N_Cam_T_pasos(N,T)
    return zeros(Int32,N,T)
end

function paso(x)
    for i in 1:length(x)
        δ=rand([-1,1])
        x[i] += δ
    end
    return x
end

function caminatas!(x0)
    xt=copy(x0[:,1])
    xf=copy(x0)
    for t in 2:length(x0[1,:])
        xf[:,t]=paso(xt)
        xt= copy(xf[:,t])
    end
    return xf
end
```

```julia
posiciones=caminatas!(N_Cam_T_pasos(100,1000))
```

```julia
using Plots
plotly()
```

```julia
plot(transpose(posiciones[1:20:100,:]))
```

## Promedio de vectores

Calcular la distancia promedio de todos los caminantes respecto al origen

```julia
function prom_vec(v)
    suma=0
    for i in 1:length(v)
        suma += v[i]
        #println(suma, " ", v[i])
    end
    return suma/length(v)
end
```

```julia
a=[1,1,1]
prom_vec(a)
```

Queremos calcular la distancia promedio de los caminantes, desde el paso 0, hasta el paso t

```julia
function prom_todos_t(v,t)
    tiempo=[]
    for i in 1:t
        push!(tiempo,prom_vec(v[:,i]))
    end
    return tiempo
end
```

```julia
promedios_t=prom_todos_t(posiciones,1000)
```

```julia
N_cam=100
N_pasos=1000
posiciones=caminatas!(N_Cam_T_pasos(N_cam,N_pasos))
promedios_t=prom_todos_t(posiciones,N_pasos)
plot(promedios_t)
```

```julia
ale=rand(100)
histogram(ale,bins=0:0.1:1,normed=true)
```

```julia
dist=posiciones[:,10]
cajas=collect((minimum(dist):maximum(dist)))
histogram(dist,bins=cajas)
```
